This notebook contains the two required exercises:

1. Sort an Array of 0s, 1s, and 2s (OOP)
2. Binary Search (Functional Programming)


```python
class Solution:
    def sort012(self, arr, n):
        low = 0
        mid = 0
        high = n - 1

        while mid <= high:
            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
            elif arr[mid] == 1:
                mid += 1
            else:
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1

        return arr

sol = Solution()
sol.sort012([0, 2, 1, 2, 0], 5)
```

```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        compare = lambda x, y: x - y

        if compare(arr[mid], target) == 0:
            return mid
        elif compare(arr[mid], target) < 0:
            low = mid + 1
        else:
            high = mid - 1

    return -1

binary_search([1, 3, 5, 7, 9], 7)
```
